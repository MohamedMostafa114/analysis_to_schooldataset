# Exploring Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load your dataset
df = pd.read_csv("schools.csv")
df.head()
```

```python
df.info()
```

# Data Cleaning

```python
df['total_score'] = df['average_math'] + df['average_reading'] + df['average_writing']
```

```python
# Fill missing numeric values with mean
numeric_columns = ['average_math', 'average_reading', 'average_writing', 'percent_tested','total_score']
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())

# Fill missing text values with 'Unknown'
text_columns = ['school_name', 'borough','building_code']
df[text_columns] = df[text_columns].fillna('Unknown')
```

```python
df['average_math'] = pd.to_numeric(df['average_math'], errors='coerce')
df['average_reading'] = pd.to_numeric(df['average_reading'], errors='coerce')
df['average_writing'] = pd.to_numeric(df['average_writing'], errors='coerce')
df['percent_tested'] = pd.to_numeric(df['percent_tested'], errors='coerce')
```

```python
df = df.drop_duplicates()
```

```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df

# Apply outlier removal to numeric columns
for column in numeric_columns:
    df = remove_outliers(df, column)
```

# Visualizations

```python
fig = plt.figure(figsize=(15, 10))
# Histogram for numerical columns
for i, col in enumerate(numeric_columns, 1):
    plt.subplot(2, 3, i)
    sns.histplot(data=df, x=col, kde=True)
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
correlation_matrix = df[numeric_columns].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Average scores by borough
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='borough', y='average_reading')
plt.title('Average Reading Scores by Borough')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='average_reading', y='average_writing')
plt.title('Reading vs Writing Scores')
plt.show()
```

```python
# Clean and convert score columns
score_cols = ["average_math", "average_reading", "average_writing"]
for col in score_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# Drop rows with missing values in score columns
df_clean = df.dropna(subset=score_cols).copy()

# Calculate average total score
df_clean["average_total"] = df_clean[score_cols].mean(axis=1)

# Top 10 schools
top10 = df_clean.nlargest(10, "average_total")
plt.figure(figsize=(10, 6))
sns.barplot(x="average_total", y="school_name", data=top10, palette="Greens_d")
plt.title("Top 10 Schools by Average Total Score")
plt.xlabel("Average Score")
plt.ylabel("School")
plt.tight_layout()
plt.show()

# Bottom 10 schools
bottom10 = df_clean.nsmallest(10, "average_total")
plt.figure(figsize=(10, 6))
sns.barplot(x="average_total", y="school_name", data=bottom10, palette="Reds_d")
plt.title("Bottom 10 Schools by Average Total Score")
plt.xlabel("Average Score")
plt.ylabel("School")
plt.tight_layout()
plt.show()
```

```python
# Calculate average total score by borough
avg_by_borough = df.groupby('borough')['total_score'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=avg_by_borough.index, 
            y=avg_by_borough.values,
            palette='viridis')
plt.title('Average Total Score by Borough', pad=20, fontsize=14)
plt.xlabel('Borough', fontsize=12)
plt.ylabel('Average Total Score', fontsize=12)
plt.xticks(rotation=45)

# Add value labels on top of bars
for i, v in enumerate(avg_by_borough.values):
    plt.text(i, v, f'{v:.1f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

# Create performance categories based on total score

```python
df['performance_category'] = pd.qcut(df['total_score'], q=4, 
                                   labels=['Low', 'Medium', 'High', 'Very High'])
```

```python
performance_stats = df.groupby('performance_category')[
    ['total_score', 'average_math', 'average_reading', 'average_writing']
].mean().round(2)
print(performance_stats)
```
